```python
# Calculate the regular count from cnt column
RCount=merged_df.groupby('dteday')['cnt'].sum().reset_index()

plt.figure(figsize=(14,6))
sns.lineplot(x='dteday',y='cnt', data=RCount, marker= 'o', color='black')

for x,y in zip(RCount['dteday'],RCount['cnt']):
    plt.text(x,y, str(y), fontsize=10, color='blue')

plt.title('Total Bike Rentals per day')
plt.xlabel('Date')
plt.ylabel('Total count')
plt.grid()
```

```python
# CAlculate total count from cnt column with each weekday
wcount=merged_df.groupby('weekday')['cnt'].sum().reset_index()

wlabels=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

# Apply the lambda function to map the weekday indices to their labels
wcount['weekday'] = wcount['weekday'].apply(lambda x: wlabels[x])
wcount=wcount.set_index('weekday').reindex(wlabels).reset_index()

plt.figure(figsize=(10,5))
sns.barplot(x='weekday', y='cnt', data=wcount, palette='Reds')

for index, value in enumerate(wcount['cnt']):
    plt.text(index, value, str(value), fontsize=10)

plt.xlabel('weekday')
plt.xlabel('Regular count')
plt.title('Total Bike Rentals Per Weekday')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
Final_Dataset[['weekday','cnt','mnth']].info()
```

```python
# Define month mapping
months = {1: 'January', 2: 'February'}

# Define weekday order
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Ensure weekday column is ordered
Final_Dataset['weekday'] = pd.Categorical(Final_Dataset['weekday'], categories=weekday_order, ordered=True)

# Sort by weekday to ensure order in plot
Final_Dataset = Final_Dataset.sort_values(by='weekday')

# Create the bar plot
plt.figure(figsize=(14, 6))
l = sns.barplot(data=Final_Dataset, x='weekday', y='cnt', hue='mnth', palette='dark', estimator=sum)

# Adjust legend labels (mapping numbers to month names)
handles, labels = l.get_legend_handles_labels()

# Convert numeric labels to month names
#labels = [months[int(label)] for label in labels]
labels = [months[int(float(label))] for label in labels]
# Update the legend with correct month names
l.legend(handles, labels, title='Month')

# Labeling and formatting
plt.xlabel('Weekday')
plt.ylabel('Total Rentals')
plt.title('Total Rentals per Weekday for January & February')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
